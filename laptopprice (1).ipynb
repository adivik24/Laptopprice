```python
import re

# Extract processor brand (e.g., "Intel", "AMD", "Apple")
def extract_processor_brand(proc):
    if proc.startswith("Apple"):
        return "Apple"
    elif "Intel" in proc:
        return "Intel"
    elif "AMD" in proc:
        return "AMD"
    return "Unknown"  # Fallback for unexpected cases

# Extract series (e.g., "Core i7", "Ryzen 5", "M1")
def extract_series(proc):
    match = re.search(r'(Intel Core|AMD Ryzen|Apple M\d)', proc)
    return match.group(1) if match else proc.split()[0]  # Handle cases like Apple M1

# Apply fixes
df['processor_brand2'] = df['processor'].apply(extract_processor_brand)
df['series'] = df['processor'].apply(extract_series)
```

```python
import re

# Extract total core count
def extract_total_cores(core_info):
    match = re.search(r'(\d+) Cores?', core_info)  # Matches '12 Cores', '14 Cores'
    if not match:
        match = re.search(r'(\w+) Core', core_info)  # Matches 'Hexa Core', 'Octa Core'
        core_map = {"Dual": 2, "Quad": 4, "Hexa": 6, "Octa": 8, "Deca": 10}
        return core_map.get(match.group(1), "Unknown") if match else "Unknown"
    return int(match.group(1))

# Extract total threads
def extract_threads(core_info):
    match = re.search(r'(\d+) Threads?', core_info)
    return int(match.group(1)) if match else "Unknown"

# Extract performance (P) and efficiency (E) cores
def extract_p_e_cores(core_info):
    match = re.search(r'(\d+)P \+ (\d+)E', core_info)
    return (int(match.group(1)), int(match.group(2))) if match else ("Unknown", "Unknown")

# Apply functions to DataFrame
df['total_cores'] = df['CPU'].apply(extract_total_cores)
df['total_threads'] = df['CPU'].apply(extract_threads)
df[['performance_cores', 'efficiency_cores']] = df['CPU'].apply(lambda x: pd.Series(extract_p_e_cores(x)))
```

```python
import pandas as pd
import re

# Sample GPU data (Replace this with your actual DataFrame column)
gpu_list=df["GPU"]

# Function to clean GPU column
def process_gpu(gpu_name):
    if pd.isna(gpu_name) or gpu_name.strip() == "":
        return "Unknown", "Unknown", 0  # If missing, return Unknown
    
    gpu_name = gpu_name.strip().replace("  ", " ")  # Remove extra spaces

    # Extract VRAM size (if available)
    vram_match = re.search(r'(\d+)GB', gpu_name)
    vram = int(vram_match.group(1)) if vram_match else 0  # Set to 0 if not found

    # Identify GPU Brand
    if "NVIDIA" in gpu_name:
        brand = "NVIDIA"
    elif "AMD" in gpu_name:
        brand = "AMD"
    elif "Intel" in gpu_name:
        brand = "Intel"
    else:
        brand = "Unknown"

    # Extract GPU Model
    model = re.sub(r'\d+GB', '', gpu_name)  # Remove VRAM part
    model = model.replace(brand, "").strip()  # Remove brand name
    model = re.sub(r'Graphics|Integrated', '', model, flags=re.I).strip()  # Remove unnecessary words

    return brand, model, vram

# Convert list to DataFrame
t = pd.DataFrame(df,columns=["GPU"])

# Apply processing function
t[['GPU_Brand', 'GPU_Model', 'GPU_VRAM']] = t['GPU'].apply(lambda x: pd.Series(process_gpu(x)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor

def train_model(df, target_col="price", test_size=0.29, random_state=42, **model_params):
    """
    Trains a specified regression model on the given DataFrame, evaluates its performance, and returns the model along with errors.
    
    Parameters:
    df (pd.DataFrame): The dataset containing features and target.
    model_class: The regression model class (e.g., RandomForestRegressor, XGBRegressor, etc.).
    target_col (str): The name of the target column.
    test_size (float): The proportion of data to use as the test set.
    random_state (int): Random seed for reproducibility.
    **model_params: Additional hyperparameters for the chosen model.

    Returns:
    dict: Dictionary containing the trained model and error metrics.
    """
    # Split dataset
    model_class= RandomForestRegressor
    X = df.drop(columns=[target_col])  # Features
    y = df[target_col]  # Target

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Initialize and train the model
    model = model_class(random_state=random_state, **model_params)
    model.fit(X_train, y_train)

    # Get predictions
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Calculate errors
    metrics = {
        "Train MSE": mean_squared_error(y_train, y_pred_train),
        "Test MSE": mean_squared_error(y_test, y_pred_test),
        "Train RMSE": mean_squared_error(y_train, y_pred_train) ** 0.5,  # Fixed calculation
        "Test RMSE": mean_squared_error(y_test, y_pred_test) ** 0.5,  # Fixed calculation
        "Train MAE": mean_absolute_error(y_train, y_pred_train),
        "Test MAE": mean_absolute_error(y_test, y_pred_test),
        "Train R2": r2_score(y_train, y_pred_train),
        "Test R2": r2_score(y_test, y_pred_test),
    }

    return {"model": model, "metrics": metrics, "X_test": X_test, "y_test": y_test, "y_pred_test": y_pred_test}
```

```python

results_rf = train_model(df,n_estimators=175,max_depth=19,random_state=42)
print(results_rf["metrics"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor  # Import XGBRegressor

def train_model_xgb(df, target_col="price", test_size=0.29, random_state=42, **model_params):
    """
    Trains a specified regression model on the given DataFrame, evaluates its performance, and returns the model along with errors.
    
    Parameters:
    df (pd.DataFrame): The dataset containing features and target.
    model_class: The regression model class (e.g., RandomForestRegressor, XGBRegressor, etc.).
    target_col (str): The name of the target column.
    test_size (float): The proportion of data to use as the test set.
    random_state (int): Random seed for reproducibility.
    **model_params: Additional hyperparameters for the chosen model.

    Returns:
    dict: Dictionary containing the trained model and error metrics.
    """
    # Split dataset
    model_class= XGBRegressor
    X = df.drop(columns=[target_col])  # Features
    y = df[target_col]  # Target

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Initialize and train the model
    model = model_class(random_state=random_state, **model_params)
    model.fit(X_train, y_train)

    # Get predictions
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Calculate errors
    metrics = {
        "Train MSE": mean_squared_error(y_train, y_pred_train),
        "Test MSE": mean_squared_error(y_test, y_pred_test),
        "Train RMSE": mean_squared_error(y_train, y_pred_train) ** 0.5,  # Fixed calculation
        "Test RMSE": mean_squared_error(y_test, y_pred_test) ** 0.5,  # Fixed calculation
        "Train MAE": mean_absolute_error(y_train, y_pred_train),
        "Test MAE": mean_absolute_error(y_test, y_pred_test),
        "Train R2": r2_score(y_train, y_pred_train),
        "Test R2": r2_score(y_test, y_pred_test),
    }

    return {"model": model, "metrics": metrics, "X_test": X_test, "y_test": y_test, "y_pred_test": y_pred_test}
```

```python

results_xgb = train_model_xgb(df, max_depth=10, learning_rate=0.12,n_estimators=1000, random_state=42)
print(results_xgb["metrics"])
```
